```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import minmax_scale
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
```

```python
veri = pd.read_csv("kalp_rahatsizligi2.csv")
veri.head(3)
```

```python
veri.describe()
```

```python
veri.info()
```

```python
veri.rename(columns={
"gogus_agrisi_tipi":"gat",
"hareketsiz_kan_basinci":"hkb",
"serum_kolestrol":"sk",
"aclik_kan_sekeri":"aks",
"elektrokardiyografi":"ekg",
"en_yuksek_kalp_hizi":"eykh",
"anjin_bagli_egsersiz":"abe",
"st_depresyonu":"st_d",
"st_egimi":"st_e",
"buyuk_damarlar":"bds"},inplace=True)
```

```python
veri.info()
```

```python
#kayıp veri olan sütunları silmek axis=1--> sütunlar axis=0 satrılar
veri2=veri.drop(labels=["bds","talasemi"],axis=1)
veri2.info()
# veri dataframe değişmedi
```

```python
# eksik ver olan satırları sileceğim için axis=0
veri3=veri.dropna(axis=0)
veri3.info()
```

```python
# sayısal verilerde eksik verileri ortanca ile değiştirecek işlem
imp_sayisal=SimpleImputer(missing_values=np.nan,strategy="median")# imputer tanımnladık
# kategorik verilerde eksik verileri ençok tekrar eden veri ile değiştirecek işlem
imp_kat=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
```

```python
veri4=veri.copy()
#veri4.info()
bds=veri4["bds"].values.reshape(-1,1)
tal=veri4["talasemi"].values.reshape(-1,1)
veri4["bds"]=imp_sayisal.fit_transform(bds)
veri4["talasemi"]=imp_kat.fit_transform(tal)
veri4.info()
#veri4 dataframe inde eksik satısal veriler ortanca ile eksik kategorik veriler en çok tekrar eden değerle değiştirilid.
```

```python
print(veri4.describe()["sk"]["25%"])
print(veri4.describe()["sk"]["75%"])
```

```python
# veri4 üstünde bulunan sk(serum_kolestrol) değerinde bulunan gürültülü/aykırı veriler temizleniyor
sk_q1=veri4["sk"].describe()["25%"]
sk_q3=veri4["sk"].describe()["75%"]
sk_IQR=sk_q3-sk_q1
alt_sinir=sk_q1-1.5*sk_IQR
ust_sinir=sk_q3+1.5*sk_IQR
veri4=veri4[(veri4["sk"]>alt_sinir)&(veri4["sk"]<ust_sinir)]
veri4.info()
```

```python
# bds(11),talasemi(12) sütunlarının 50-60 arası satır bilgisini getiriniz.
veri4.info()
veri4.iloc[50:60,11:13]
```

```python
# yas, hkb,bds sütunlarına ait 20,25,30 nolu satırları görüntüle
veri4.info()
veri4.iloc[[20,25,30],[0,3,11]]
```

```python
# sadece sayısal veri içerin sütunları ve tüm satırlarını alalım ve 0-1 arasına minmax_scale fonskiyonu ile normalize edelim
veri4.iloc[:,[0,3,4,7,9,11]]=minmax_scale(X=veri4.iloc[:,[0,3,4,7,9,11]].values,feature_range=(0,1))
veri4.iloc[:,[0,3,4,7,9,11]]
```

```python
veri4.head(3)
```

```python
le=LabelEncoder()
veri4["cinsiyet"]=le.fit_transform(y=veri4["cinsiyet"])
veri4["gat"]=le.fit_transform(y=veri4["gat"])
veri4["ekg"]=le.fit_transform(y=veri4["ekg"])
veri4["st_e"]=le.fit_transform(y=veri4["st_e"])
veri4["talasemi"]=le.fit_transform(y=veri4["talasemi"])
veri4.head(3)
```

```python
veri4.info()
```

```python
veri4.to_csv("kalp_rahatsizligi_temiz.csv",index=False)
```
