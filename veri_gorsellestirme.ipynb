```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
veri=pd.read_csv("kalp_rahatsizligi.csv")
veri.head(3)
```

```python
sns.histplot(data=veri,x="yas",bins=10)
plt.xlabel("yaş")
plt.ylabel("adet/frekans")
plt.show()
```

```python
# serum kolestrol değerinin histogram grafiğini çizdiriniz.
# veri.info()
sns.histplot(data=veri,x="serum_kolestrol")
plt.xlabel("serum kolestrol mg/dl")
plt.ylabel("adet/frekans")
plt.show()
```

```python
sns.histplot(data=veri,x="serum_kolestrol",hue="aclik_kan_sekeri",element="step")
plt.xlabel("serum kolestrol mg/dl")
plt.ylabel("adet/frekans")
plt.show()
```

```python
# yaş sayısal verisinin göğüs ağrısı tipine göre histogram grafiğini çizdirip yorumlayınız.
#gogus_agrisi_tipi
sns.histplot(data=veri,x="yas",hue="gogus_agrisi_tipi",element="step")
plt.xlabel("yaş/göğüs ağrısı tipine göre")
plt.ylabel("adet/frekans")
plt.show()
```

### saçılım(scatter) grafiği
* iki sayısal veri arasındaki ilişkiyi gözlemlememizi sağlar


```python
sns.scatterplot(data=veri,x="yas",y="serum_kolestrol")
plt.show()
```

```python
sns.lmplot(data=veri,x="yas",y="serum_kolestrol")
plt.show()
```

```python
sns.lmplot(data=veri,x="yas",y="en_yuksek_kalp_hizi",hue="cinsiyet")
plt.show()
```

```python
sns.boxplot(data=veri,y="serum_kolestrol")
plt.show()
```
